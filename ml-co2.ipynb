```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

# Load your dataset
# Assuming your dataset is in a CSV file named 'co2_emissions.csv' with columns 'Year', 'Country', and 'CO2_Emissions'
data = pd.read_csv('co2_emissions.csv')

# Assuming you have already preprocessed your data and have it in a suitable format for training

# List of unique countries in your dataset
countries = data['Country'].unique()

# DataFrame to store predictions
submission_df = pd.DataFrame(columns=['Country', 'Year', 'CO2_Emissions'])

# Train and predict for each country separately
for country in countries:
    country_data = data[data['Country'] == country]
    X = country_data[['Year']]  # Assuming you have more features, include them accordingly
    y = country_data['CO2_Emissions']
    
    # Create and train the model
    model = LinearRegression()
    model.fit(X, y)
    
    # Make predictions for the next 10 years
    future_years = np.array([2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2030]).reshape(-1, 1)  # Years you want to predict for
    future_predictions = model.predict(future_years)
    
    # Append predictions to submission DataFrame
    country_predictions_df = pd.DataFrame({'Country': [country]*10, 'Year': future_years.flatten(), 'CO2_Emissions': future_predictions})
    submission_df = pd.concat([submission_df, country_predictions_df], ignore_index=True)

# Export submission DataFrame to CSV
submission_df.to_csv('co2_emissions_predictions.csv', index=False)
```
