```python
specific_string = 'CO2 emissions (metric tons per capita)'
filtered_rows = Train[Train['Indicator'] == specific_string]
```

```python

filtered_rows.drop(columns=['Indicator'], inplace=True)
filtered_rows
```

```python
melted_df = filtered_rows.melt(id_vars='Country Name', var_name='Year', value_name='CO2_Emissions')
```

```python
melted_df
```

```python

sorted_df = melted_df.groupby('Country Name').apply(lambda x: x.sort_values('Year')).reset_index(drop=True)
```

```python
sorted_df.head(20)
```

```python
import re
def remove_brackets(text):
    return re.sub(r"\[.*?\]", "", text)

for index, row in sorted_df.iterrows():
    sorted_df.at[index, 'Year'] = remove_brackets(row['Year'])
print(sorted_df)
```

```python
sorted_df
```

```python
sorted_df['Year'] = sorted_df['Year'].str.strip()
sorted_df['Year'] = pd.to_datetime(sorted_df['Year'], format='%Y')
```

```python
sorted_df
```

```python
specific_string = 'Afghanistan'
afg = sorted_df[sorted_df['Country Name'] == specific_string]
```

```python
afg.drop(columns=['Country Name'], inplace=True)
```

```python
afg
```

```python
country_data = afg['CO2_Emissions'].astype(float)
```

```python
country_data
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(country_data, order=(1,1,0))  # Adjust order as needed
model_fit = model.fit()

# Make forecast for the next 10 years
forecast = model_fit.forecast(steps=6)

# Print the forecast
#print("Forecast for CO2 emissions in", country_name)
print(forecast)


```
