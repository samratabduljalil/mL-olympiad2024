```python
import pandas as pd


# Replace 'filename.csv' with the path to your CSV file
df = pd.read_csv('/kaggle/input/mlolympiadbd2024/train.csv')
```

```python
columns_to_drop = ['WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','Date']  # List of column names to drop
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head()
```

```python
df.replace(to_replace={'RainToday': {'No': '0', 'Yes': '1'}}, inplace=True)
```

```python
df.head()
```

```python
columns_to_drop = ['ID']  # List of column names to drop
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head()
```

```python
Y_train=df['RainToday']
```

```python
Y_train
```

```python
X_train=df.drop(['RainToday'],axis=1)
     
```

```python
X_train
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
# Replace 'filename.csv' with the path to your CSV file
dftest2 = pd.read_csv('/kaggle/input/mlolympiadbd2024/test.csv')


```

```python
columns_to_drop = ['WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','Date','ID']  # List of column names to drop
dftest2.drop(columns=columns_to_drop, inplace=True)
```

```python
dftest2
```

```python
y_pred = svc.predict(dftest2)
```

```python
y_pred
```

```python
results = pd.DataFrame({
    'ID': range(1, len(dftest2) + 1),
    'Predicted': y_pred
})
```

```python
results
```

```python
results.replace(to_replace={'Predicted': {'0': 'No', '1': 'Yes'}}, inplace=True)
```

```python
results
```

```python
results.to_csv('subset.csv', index=False)
```
