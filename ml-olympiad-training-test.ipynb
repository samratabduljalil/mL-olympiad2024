```python
import pandas as pd


# Replace 'filename.csv' with the path to your CSV file
df = pd.read_csv('/kaggle/input/mlolympiadbd2024/train.csv')
```

```python

count_of_yes = (df['RainToday'] == 'Yes').sum()

print("Count of 'yes' in column 'RainToday':", count_of_yes)
```

```python

count_of_yes = (df['RainToday'] == 'No').sum()

print("Count of 'yes' in column 'RainToday':", count_of_yes)
```

```python
columns_to_drop = ['WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','Date']  # List of column names to drop
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df
```

```python
df.replace(to_replace={'RainToday': {'No': '0', 'Yes': '1'}}, inplace=True)
```

```python
df.head()
```

```python
columns_to_drop = ['ID']  # List of column names to drop
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head()
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

# Assuming you have a DataFrame named df with features X and target variable y
X = df.drop(columns=['RainToday'])
y = df['RainToday']

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)
```

```python
X_balanced.head()
```

```python
# Replace 'filename.csv' with the path to your CSV file
dftest2 = pd.read_csv('/kaggle/input/mlolympiadbd2024/test.csv')


```

```python
columns_to_drop = ['WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','Date','ID']  # List of column names to drop
dftest2.drop(columns=columns_to_drop, inplace=True)
```

```python
dftest2.head(10)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Load your prepared training and testing data
# X_train, y_train, X_test are assumed to be already prepared

# Reshape input data for LSTM (assuming X_train and X_test are 2D arrays)
# X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
# X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_balanced.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_balanced.astype(float), y_balanced.astype(float), epochs=25, batch_size=31)

# Make predictions on the test data
predictions = model.predict(dftest2)

# You may want to inverse transform the predictions if you scaled your data
# scaler.inverse_transform(predictions)
```

```python
binary_predictions = np.where(predictions.ravel() > 0.5, 1, 0)
```

```python
binary_predictions
```

```python
results = pd.DataFrame({
    'ID': range(1, len(dftest2) + 1),
    'Predicted': binary_predictions
})
```

```python
results.replace(to_replace={'Predicted': {0: 'No', 1: 'Yes'}}, inplace=True)
```

```python
results.to_csv('submission.csv', index=False)
```
